## Creating columns base on conditions
1. Price>17 --> cheap
2. 17-22 --> Moderate
3. +22 --> Expensive

```python
A = S[['Manufacturer','Model','Price']]
A.head()
```

```python
pr_label=[]
for i in S['Price']:
    if i<17 :
        pr_label.append('Cheap')
    elif i>=17 and i<22:
        pr_label.append('Moderate')
    else:
        pr_label.append('Expensive')
        
        
```

```python
print(pr_label)
```

```python
A['Price_label']=pr_label
A
```

```python
A['Price_label'].value_counts()
```

## Different Types Of Joins

```python
emp_data=pd.read_csv('emp_2.csv')
dept=pd.read_csv('dept.csv')
```

```python
emp_data
```

```python
dept
```

```python
inner=pd.merge(emp_data,dept,how='inner',left_on='did',right_on='dept_id')
inner
```

## Left Join

```python
left=pd.merge(emp_data,dept,how='left',left_on='did',right_on='dept_id')
left
```

## Right Join

```python
right=pd.merge(emp_data,dept,how='right',left_on='did',right_on='dept_id')
right
```

## Outer Join

```python
outer=pd.merge(emp_data,dept,how='outer',left_on='did',right_on='dept_id')
outer
```
